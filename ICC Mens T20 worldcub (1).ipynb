**03.What are the common decisions made by teams after winning the toss?**

```python
import matplotlib.pyplot as plt
import seaborn as sns

pivot_table = pd.pivot_table(df, index='Toss Winning', columns='Toss Decision', values='1st Team', aggfunc="count")

# Using matplotlib
pivot_table.plot(kind='bar')
plt.title('Toss Decisions by Winning Team')
plt.xlabel('Toss Winning Team')
plt.ylabel('Count')
plt.show()

# Using seaborn
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu')
plt.title('Toss Decisions by Winning Team')
plt.show()
```

### **Weather Impact:**

**01.How many matches were affected by rain?**

```python
rain_affected_matches = df[df['Winning Margin'].str.contains('Rain', na=False)]
num_rain_affected_matches = len(rain_affected_matches)
print(num_rain_affected_matches)
```

**02.What is the impact of rain on match outcomes and player performances?**

```python
# Filter for matches where the first team batting won
first_batting_wins = rain_affected_matches[rain_affected_matches['1st Team'].str.contains('runs', na=False)]

# Filter for matches where the second team batting won
second_batting_wins = rain_affected_matches[rain_affected_matches['2nd Team'].str.contains('wickets', na=False)]

# Calculate and print the percentages
percentage_first_batting_wins = (len(first_batting_wins) / num_rain_affected_matches) * 100
percentage_second_batting_wins = (len(second_batting_wins) / num_rain_affected_matches) * 100

print("Percentage of rain-affected matches won by the team batting first: {:.2f}%".format(percentage_first_batting_wins))
print("Percentage of rain-affected matches won by the team batting second: {:.2f}%".format(percentage_second_batting_wins))
```
